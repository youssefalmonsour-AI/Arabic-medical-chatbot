```python
import pandas as pd
import re
```

```python
def clean_text(text):
    text = re.sub(r'[\u064b-\u065f]', '', text)  # إزالة التشكيل
    text = re.sub(r'[^\u0600-\u06FF\s؟]', '', text)  # إزالة الرموز
    return text.strip()
```

```python
df = pd.read_csv("../data/prepared_medical_data.csv").dropna()
```

```python
df["Question"] = df["Question"].astype(str).apply(clean_text)
df["Answer"] = df["Answer"].astype(str).apply(clean_text)
```

```python
df["input"] = "سؤال: " + df["Question"]
df["output"] = "إجابة: " + df["Answer"]
```

```python
df[["input", "output"]].to_csv("../data/cleaned_for_generation.csv", index=False)
print(" تم تجهيز البيانات بنجاح.")
```
